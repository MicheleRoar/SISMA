```python
import ast
import re
import pandas as pd

nap_set = {str(a).strip().lower() for a in nap if str(a).strip()}

_split_re = re.compile(
    r"\s*;\s*|\s*,\s*|\s*/\s*|\s*&\s*|\s+feat\.?\s+|\s+featuring\s+",
    flags=re.IGNORECASE,
)

def parse_artists_cell(x) -> list[str]:
    if x is None or (isinstance(x, float) and pd.isna(x)):
        return []
    s = str(x).strip()
    if not s:
        return []

    # caso: "['A', 'B']"
    if s.startswith("[") and s.endswith("]"):
        try:
            out = ast.literal_eval(s)
            if isinstance(out, list):
                return [str(v).strip().lower() for v in out if str(v).strip()]
        except Exception:
            pass

    # fallback: "A; B", "A, B", "A feat. B" ...
    parts = _split_re.split(s)
    return [p.strip().lower() for p in parts if p and p.strip()]

mask = tracks["artists"].apply(lambda x: any(a in nap_set for a in parse_artists_cell(x)))
out = tracks[mask].copy()
```

```python
from collections import Counter
import pandas as pd
import ast

# -------------------------
# helpers
# -------------------------
def parse_list(x):
    if pd.isna(x):
        return []
    if isinstance(x, list):
        return x
    try:
        return ast.literal_eval(x)
    except Exception:
        return []

tracks["id_artists_list"] = tracks["id_artists"].apply(parse_list)
artists["genres_list"] = artists["genres"].apply(parse_list)

artist_to_genres = dict(
    zip(artists["id"].astype(str), artists["genres_list"])
)

# -------------------------
# HARD perceptual filter
# -------------------------
def filter_soft_preset(df, preset):
    x = df.copy()

    if not preset.get("allow_explicit", False):
        x = x[x["explicit"] == 0]

    for k in ["tempo", "energy", "danceability", "valence", "loudness"]:
        mn = preset.get(f"{k}_min")
        mx = preset.get(f"{k}_max")
        if mn is not None:
            x = x[x[k] >= mn]
        if mx is not None:
            x = x[x[k] <= mx]

    return x


# -------------------------
# genre extraction (SOFT)
# -------------------------
rows = []

for preset_name, preset in PRESETS.items():
    df_f = filter_soft_preset(tracks, preset)

    genre_counter = Counter()

    for _, row in df_f.iterrows():
        genres = set()
        for aid in row["id_artists_list"]:
            genres.update(artist_to_genres.get(str(aid), []))

        if not genres:
            continue

        weight = 1.0 / len(genres)
        for g in genres:
            genre_counter[g] += weight

    if not genre_counter:
        continue

    total = sum(genre_counter.values())
    top20 = genre_counter.most_common(20)

    for genre, w in top20:
        rows.append({
            "preset": preset_name,
            "genre": genre,
            "weight": round(w / total, 3)
        })

# -------------------------
# final table
# -------------------------
top_genres_soft = (
    pd.DataFrame(rows)
      .sort_values(["preset", "weight"], ascending=[True, False])
      .reset_index(drop=True)
)
```

# Associare generi musicali ad una nazione

```python
import ast
import re
from collections import Counter
import pandas as pd
import math

# -----------------------------
# helpers
# -----------------------------
def parse_genres(x):
    if x is None:
        return []
    if isinstance(x, list):
        return [g.lower().strip() for g in x]
    if isinstance(x, str):
        x = x.strip()
        if not x:
            return []
        if x.startswith("["):
            try:
                return [g.lower().strip() for g in ast.literal_eval(x)]
            except Exception:
                return []
        return []
    return []

def normalize(g: str) -> str:
    return re.sub(r"\s+", " ", str(g).lower().strip())

def contains_token(text: str, token: str) -> bool:
    """Match token with word boundaries (safer than raw substring for single words)."""
    return re.search(rf"\b{re.escape(token)}\b", text) is not None

# -----------------------------
# config (tua)
# -----------------------------
NATIONS = [
    "italian","french","german","spanish","portuguese",
    "british","uk","english",
    "irish","scottish","welsh",
    "albanian",
    "swedish","norwegian","danish","finnish","icelandic",
    "canadian","american",
    "dutch","belgian","austrian","swiss",
    "polish","czech","hungarian","romanian","bulgarian",
    "greek","turkish",
    "australian",
    "mexican","argentine","brazilian","chilean","colombian","peruvian","venezuelan",
    "indian","japanese","chinese"
]


GENERIC = {
    "pop","rock","indie","hip hop","rap","jazz","soul","r&b",
    "electronic","dance","house","techno","metal", "progressive rock",
    "alternative","experimental","soundtrack","classical", "symphonic rock",
    "power metal", "neo classical metal", "melodic metal"
}

MIN_FREQ = 10   # fallback (cluster / general)

# -----------------------------
# NEW: conservativeness for NATIONS expansion (patch)
# -----------------------------
TOP_N_NATION = 150          # keep top-N expanded genres per nation (by score)
MIN_LIFT_NATION = 1.6       # require some specificity vs global
NATION_MIN_RATIO = 0.005     # dynamic minimum freq = ceil(ratio * n_seed), with floor below
NATION_MIN_FLOOR = 8        # avoid killing small nations completely

# -----------------------------
# NEW: clusters (macro-aree) (tua)
# -----------------------------
CLUSTERS = {
    "usa": {
        "seeds": {
            "americana","appalachian","bluegrass","cajun",
            "nashville","memphis","new orleans",
            "dmv","bay area","chicago","detroit","atlanta","brooklyn","harlem","oakland","compton",
            "texas country","texas blues","texas rock",
        },
        "boosters": {
            "west coast","east coast","southern","midwest","texas","louisiana",
            "chicago rap","detroit hip hop","atlanta hip hop","bay area rap",
        },
        "stop": {"american pop","american rock"},
        "min_freq": 10,
        "min_lift": 2.0,          # P(g|cluster) / P(g|global)
        "min_artist_ratio": 0.01, # dynamic threshold: >=1% of seed artists (min 5)
    },
    "latin": {
        "seeds": {
            "reggaeton","bachata","salsa","cumbia","merengue","vallenato",
            "ranchera","mariachi","corridos","norteno","banda","tejano",
            "rock en espanol","urbano espanol","urbano latino",
        },
        "boosters": {
            "latin","latino","en espanol",
            "mexican","colombian","argentine","chilean","peruvian","venezuelan",
            "puerto rican","dominican","cuban","brazilian",
        },
        "stop": {"latin pop"},
        "min_freq": 10,
        "min_lift": 2.0,
        "min_artist_ratio": 0.01,
    },
    "mena": {
        "seeds": {"arabic","khaleeji","shaabi","dabke","gnawa","rai","tarab","middle eastern"},
        "boosters": {"arab","maghreb","levant","gulf","egyptian","lebanese","palestinian","syrian","moroccan","tunisian","algerian","iraqi"},
        "stop": {"oriental"},
        "min_freq": 8,
        "min_lift": 2.0,
        "min_artist_ratio": 0.01,
    },
    "african": {
        "seeds": {"afrobeat","afrobeats","amapiano","highlife","gqom","kwaito","mbaqanga","bongo flava","gengetone","kuduro"},
        "boosters": {"afropop","afro house","nigerian","ghanaian","south african","kenyan","tanzanian","ethiopian","congolese","angolan","ugandan"},
        "stop": {"world","global"},
        "min_freq": 8,
        "min_lift": 2.0,
        "min_artist_ratio": 0.01,
    },
}

def nation_match(g: str, nation: str) -> bool:
    g = normalize(g)
    nation = normalize(nation)
    # se è una sigla corta tipo "uk", usa word boundary
    if len(nation) <= 3:
        return contains_token(g, nation)
    # per parole lunghe tipo "italian", substring va bene
    return nation in g

def _cluster_seed_genres(all_genres, cfg):
    """
    Seed genres for a cluster using:
      - safer matching:
        * if seed/booster is 1 token -> word-boundary match
        * if it is multi-token -> substring match (after normalize)
      - exclusions via stop
    """
    seeds = {normalize(x) for x in cfg.get("seeds", set())}
    boosters = {normalize(x) for x in cfg.get("boosters", set())}
    stop = {normalize(x) for x in cfg.get("stop", set())}

    def match_any(gg: str, patterns: set[str]) -> bool:
        for p in patterns:
            if p in stop:
                continue
            if " " in p:  # phrase
                if p in gg:
                    return True
            else:  # single token
                if contains_token(gg, p):
                    return True
        return False

    out = set()
    for g in all_genres:
        gg = normalize(g)
        if gg in stop:
            continue
        if match_any(gg, seeds) or match_any(gg, boosters):
            out.add(gg)
    return out

# -----------------------------
# preprocess
# -----------------------------
artists = artists.copy()
artists["genres_parsed"] = artists["genres"].apply(parse_genres)
artists["genres_parsed"] = artists["genres_parsed"].apply(lambda gs: [normalize(g) for g in gs])

all_genres = set(g for row in artists["genres_parsed"] for g in row)

def is_valid_genre(g: str) -> bool:
    if g in GENERIC:
        return False
    # regola originale: scarta 1-token
    if len(g.split()) == 1:
        return False
    return True

# Global counts PER ARTIST (presence), non per-occurrence
global_counter = Counter()
for gs in artists["genres_parsed"]:
    for g in set(gs):
        if is_valid_genre(g):
            global_counter[g] += 1

global_total_artists = len(artists)

# -----------------------------
# analysis (NATIONS + CLUSTERS)
# -----------------------------
results = {}

# ---- NATIONS (PATCH: conservativo con score + top-N + min lift) ----
for nation in NATIONS:
    seed_genres = {g for g in all_genres if nation_match(g, nation)}
    if not seed_genres:
        continue

    mask_seed = artists["genres_parsed"].apply(lambda gs: any(g in seed_genres for g in gs))
    seed_authors = artists[mask_seed]
    n_seed = len(seed_authors)
    if n_seed == 0:
        continue

    # per-artist presence counter inside nation seeds
    counter = Counter()
    for gs in seed_authors["genres_parsed"]:
        for g in set(gs):
            if is_valid_genre(g):
                counter[g] += 1

    # dynamic min freq (prevents forcing ">= 20" for tiny nations)
    min_freq_dynamic = max(NATION_MIN_FLOOR, int(math.ceil(NATION_MIN_RATIO * n_seed)))

    filtered_items = []
    nation_scores = {}      # score = c * log1p(lift)
    nation_lifts = {}       # for debug

    for g, c in counter.items():
        if c < min_freq_dynamic:
            continue

        base = global_counter.get(g, 0)
        p_in = c / max(1, n_seed)
        p_gl = max(1e-12, base / max(1, global_total_artists))
        lift = p_in / p_gl

        if lift < MIN_LIFT_NATION:
            continue

        score = c * math.log1p(lift)  # stable: log(1+lift)
        nation_scores[g] = float(score)
        nation_lifts[g] = float(lift)
        filtered_items.append((g, c, score, lift))

    # order by score (then count) and take top-N
    filtered_items.sort(key=lambda x: (x[2], x[1]), reverse=True)
    top = filtered_items[:TOP_N_NATION]

    filtered = Counter({g: c for (g, c, _, __) in top})

    results[nation] = {
        "type": "nation",
        "seed_genres": sorted(seed_genres),
        "n_seed_authors": n_seed,
        "expanded_genres": filtered,
        "lift_scores": nation_lifts,   # debug
        "score": nation_scores,        # debug
        "nation_cfg": {
            "top_n": TOP_N_NATION,
            "min_lift": MIN_LIFT_NATION,
            "min_ratio": NATION_MIN_RATIO,
            "min_floor": NATION_MIN_FLOOR,
            "min_freq_dynamic": min_freq_dynamic,
        },
    }

# ---- CLUSTERS (come tuo: per-artist + lift corretto + soglia dinamica) ----
for cname, cfg in CLUSTERS.items():
    seed_genres = _cluster_seed_genres(all_genres, cfg)
    if not seed_genres:
        continue

    mask_seed = artists["genres_parsed"].apply(lambda gs: any(g in seed_genres for g in gs))
    seed_authors = artists[mask_seed]
    n_seed = len(seed_authors)
    if n_seed == 0:
        continue

    # per-artist presence counter inside cluster
    counter = Counter()
    for gs in seed_authors["genres_parsed"]:
        for g in set(gs):
            if is_valid_genre(g):
                counter[g] += 1

    # thresholds
    min_freq_cfg = int(cfg.get("min_freq", MIN_FREQ))
    min_lift = float(cfg.get("min_lift", 2.0))
    min_artist_ratio = float(cfg.get("min_artist_ratio", 0.01))
    min_freq_dynamic = max(5, int(math.ceil(min_artist_ratio * n_seed)))
    min_freq = max(min_freq_cfg, min_freq_dynamic)

    filtered = Counter()
    scores = {}  # lift scores (debug)

    for g, c in counter.items():
        if c < min_freq:
            continue

        base = global_counter.get(g, 0)
        p_in = c / max(1, n_seed)
        p_gl = max(1e-12, base / max(1, global_total_artists))

        lift = p_in / p_gl
        if lift >= min_lift:
            filtered[g] = c
            scores[g] = float(lift)

    results[cname] = {
        "type": "cluster",
        "seed_genres": sorted(seed_genres),
        "n_seed_authors": n_seed,
        "expanded_genres": filtered,
        "lift_scores": scores,  # utile per debug/ordinamento
        "cluster_cfg": {k: list(v) if isinstance(v, set) else v for k, v in cfg.items()},
        "thresholds": {
            "min_freq_cfg": min_freq_cfg,
            "min_artist_ratio": min_artist_ratio,
            "min_freq_dynamic": min_freq_dynamic,
            "min_freq": min_freq,
            "min_lift": min_lift,
        },
    }
```

```python
import json
from collections import Counter

def _to_jsonable(obj):
    if isinstance(obj, Counter):
        return dict(obj)
    if isinstance(obj, dict):
        return {k: _to_jsonable(v) for k, v in obj.items()}
    if isinstance(obj, list):
        return [_to_jsonable(x) for x in obj]
    return obj

out = _to_jsonable(results)

with open("data/regions_results.json", "w", encoding="utf-8") as f:
    json.dump(out, f, ensure_ascii=False, indent=2)
```

```python
import numpy as np
import pandas as pd

from app.services.data_store import DataStore
from app.services.recommender import primary_artist_from_field
from app.services.region_genres import get_region_payload

# ----------------------------
# 1) Load
# ----------------------------
store = DataStore(
    tracks_csv_path="data/archive/tracks.csv",
    artists_csv_path="data/archive/artists.csv",
)
df = store.get_df()

print("TOTAL TRACKS:", len(df))

# ----------------------------
# 2) REGION IT universe (genres from payload)
# ----------------------------
it_payload = get_region_payload(iso="IT", top_n=5000)
assert it_payload.get("ok"), it_payload

it_genres = [g.lower().strip() for g in it_payload["genres"]]
idx_it = np.asarray(store.get_row_indices_by_genres(it_genres), dtype=np.int64)

print("\nREGION IT GENRES:", len(it_genres))
print("UNIVERSE IT TRACKS:", idx_it.size)

# sanity check: quanti generi realmente presenti nel DF intersecano la lista
if "genres_list" in df.columns:
    present = set(g for row in df["genres_list"] for g in (row or []))
    inter = len(set(it_genres) & set(present))
    print("IT genres present in DF:", inter)

# ----------------------------
# 3) Apply filters step-by-step
# ----------------------------
def apply_range(idx, col, mn=None, mx=None):
    vals = pd.to_numeric(df.iloc[idx][col], errors="coerce").to_numpy(dtype=float)
    m = np.ones(idx.size, dtype=bool)
    if mn is not None:
        m &= vals >= float(mn)
    if mx is not None:
        m &= vals <= float(mx)
    return idx[m]

def report(label, idx):
    print(f"{label:30s}: {idx.size}")

ranges = {
    "tempo": (70.0, 90.0),
    "danceability": (0.35, 0.55),
    "energy": (0.45, 0.65),
    "valence": (0.65, 0.85),
    "loudness": (-16.0, -10.0),
}

idx = idx_it.copy()
print("\n--- STEP COUNTS ---")
report("IT universe", idx)

# explicit hard (se vuoi simulare Discovery)
if "explicit" in df.columns:
    exp = pd.to_numeric(df["explicit"], errors="coerce").fillna(0).astype(int).to_numpy()
    idx = idx[exp[idx] == 0]
    report("after explicit=0", idx)

for col, (mn, mx) in ranges.items():
    idx = apply_range(idx, col, mn, mx)
    report(f"after {col} [{mn},{mx}]", idx)

# ----------------------------
# 4) max_per_artist impact
# ----------------------------
max_per_artist = 2
artists = df.iloc[idx]["artists"].apply(primary_artist_from_field).astype(str).fillna("").to_numpy()
vc = pd.Series(artists).value_counts()
selectable = int(vc.clip(upper=max_per_artist).sum())

print("\n--- ARTIST CAP IMPACT ---")
print("Unique primary artists:", int((vc.index != "").sum()))
print("Tracks in filtered set:", int(idx.size))
print(f"Selectable with max_per_artist={max_per_artist}:", selectable)
print("\nTop artists by count:")
print(vc.head(20))
```
